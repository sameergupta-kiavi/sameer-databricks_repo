```
s3_data_location = dbutils.widgets.get("s3_data_location")
delta_table_location = dbutils.widgets.get("delta_table_location")
primary_keys = dbutils.widgets.get("primary_keys")
timestamp_keys = dbutils.widgets.get("timestamp_keys")
```

```
import pyspark.sql.functions as F
```

```
df = spark.read.format("parquet").option("header",True).load(s3_data_location)
```

```
df = df.dropDuplicates([primary_keys, timestamp_keys])
```

```
df.write.mode("overwrite").format('delta').save(delta_table_location)
```
